
# GÖREV 2: RFM Metriklerinin Hesaplanması

## recency-müşterinin yeniliği/sıcaklığı (analizin yapıldığı tarih-müşterinin son satın alma tarihi)
## Frequency- Müşterinin yaptığı toplam satın alma adet
## Monetary-Müşterinin yaptığı satın almalar sonucu bıraktığı toplam parasal değer

```python
df.head()
```

```python
rfm=df.groupby("master_id").agg({"last_order_date": lambda recency: (today_date-recency.max()).days,
                                    "Total_Ever_Order": sum,
                                     "Total_Ever_Value" : sum })
```

```python
rfm.head()
```

```python
rfm.columns=["recency","frequency","monetary"]
rfm.head(50)
```

```python
rfm.describe().T
```

```python
rfm.shape
```

# GÖREV 3: RF ve RFM Skorlarının Hesaplanması

##### Calculating RFM Scores ##skorları gzden geçir

## recency ters, monatery ve frequency düz çarpanlar getirir.

## frequency ve monetary büyükse 5 puan, recency büyükse 1 puan anlamına gelir

```python
rfm["recency_score"]=pd.qcut(rfm["recency"],5,labels=[5,4,3,2,1])
rfm.head()
```

```python
rfm["monetary_score"]=pd.qcut(rfm["monetary"],5,labels=[1,2,3,4,5])
rfm.head()
```

```python
rfm["frequency_score"]=pd.qcut(rfm["frequency"].rank(method="first"),5,labels=[1,2,3,4,5]) ##çakışmlar olduğundan hata veriyor ancak rank(method="first") kullanarak buna göre sıralamasını sağlıyoruz
rfm.head()
```

```python
rfm["RFM_SCORE"]=(rfm["recency_score"].astype(str)+rfm["frequency_score"].astype(str))
rfm.head()
```

```python
rfm.describe().T
```

```python
rfm[rfm["RFM_SCORE"]=="55"]
```

```python
rfm[rfm["RFM_SCORE"]=="11"]
```

# GÖREV 4: RF Skorlarının Segment Olarak Tanımlanması

```python
seg_map = {
        r'[1-2][1-2]': 'hibernating',
        r'[1-2][3-4]': 'at_risk',
        r'[1-2]5': 'cant_loose',
        r'3[1-2]': 'about_to_sleep',
        r'33': 'need_attention',
        r'[3-4][4-5]': 'loyal_customers',
        r'41': 'promising',
        r'51': 'new_customers',
        r'[4-5][2-3]': 'potential_loyalists',
        r'5[4-5]': 'champions'

}
```

```python
rfm["segment"]=rfm["RFM_SCORE"].replace(seg_map,regex=True)
```

```python
rfm.head()
```

```python
rfm[rfm["segment"]=="need_attention"].index
```

```python
rfm[rfm["segment"]=="cant_loose"].index
```
